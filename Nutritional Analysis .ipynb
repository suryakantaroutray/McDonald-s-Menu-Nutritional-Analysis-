```python
categories_count = dataset['Category'].value_counts().sort_index().plot(kind='barh')
categories_count.set_xlabel("Amount of Items")
categories_count.set_ylabel("Menu Categories")
```

```python
dataset[ ['Category', 'Protein'] ].groupby('Category') \
                                               .mean() \
                                               .sort_values('Protein', ascending=False) \
                                               .head(10) \
                                               .rename(columns={'protein': 'Protein G'}) \
                                               .rename_axis('Category') \
                                               .plot(kind='barh', figsize=(10,5))
```

```python
dataset[ ['Category', 'Calories'] ].groupby('Category') \
                                               .mean() \
                                               .sort_values('Calories', ascending=False) \
                                               .head(10) \
                                               .rename(columns={'calories': 'Average Calories'}) \
                                               .rename_axis('Category') \
                                               .plot(kind='barh', figsize=(10,5))
```

```python
dataset[ ['Category', 'Sodium (% Daily Value)'] ].groupby('Category') \
                                               .mean() \
                                               .sort_values('Sodium (% Daily Value)', ascending=False) \
                                               .head(10) \
                                               .rename(columns={'Sodium (% Daily Value)': 'Sodium (% Daily Value)'}) \
                                               .rename_axis('Category') \
                                               .plot(kind='barh', figsize=(10,5))
```

```python
mapping_dictionary = {'Breakfast': 0, 'Beef & Pork': 1, 'Chicken & Fish': 2}
```

```python
dataset['category_as_num'] = dataset.Category.map(mapping_dictionary)
dataset[['Category', 'category_as_num']]
```

```python
dataset.plot(kind='scatter', x='Calories from Fat', y='Calories', figsize=(20, 10), s=2, c='category_as_num', colormap='tab20b');
```

```python
mapping_dict = {'Breakfast': 0, 'Beef & Pork': 1, 'Chicken & Fish': 2, 'Salads': 3, 'Snacks & Sides': 4, 'Desserts': 5, 'Coffee & Tea': 6, 'Smoothies & Shakes': 6}
```

```python
dataset['category'] = dataset.Category.map(mapping_dict)
dataset[['Category', 'category']]
```

```python
dataset.plot(kind='scatter', x='Calories from Fat', y='Calories', figsize=(20, 10), s=2, c='category', colormap='tab20b');
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Calories', data=dataset, ci=None)
plt.title('Calorie Distribution Across Food Categories')
plt.xlabel('Food Category')
plt.ylabel('Calories')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for category in dataset['Category'].unique():
    sns.histplot(dataset[dataset['Category'] == category]['Total Fat (% Daily Value)'],
                 label=category, kde=True)
plt.title('Distribution of Total Fat (% Daily Value) Across Food Categories')
plt.xlabel('Total Fat (% Daily Value)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
highest_calories = dataset.loc[dataset['Calories'].idxmax()]
lowest_calories = dataset.loc[dataset['Calories'].idxmin()]

print("Menu item with the highest calorie count:")
print(highest_calories[['Item', 'Calories']])

print("\nMenu item with the lowest calorie count:")
print(lowest_calories[['Item', 'Calories']])
```

```python
average_dataset_category = dataset.groupby('Category')[['Total Fat (% Daily Value)', 'Carbohydrates (% Daily Value)']].mean()

print("\nAverage dataset Content of Popular Menu Categories:")
print(average_dataset_category)
```
